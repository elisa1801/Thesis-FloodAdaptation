# Data distribution
This notebook shows the data distribution for the variables that go into the clustering.

## 0. Data Preperation

### 0.1 Import and read files

```python
import seaborn as sns
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
#Reading the grouped_survey file
survey_data = pd.read_csv('../datafiles/grouped_survey.csv')
```

### 0.2 Functions

```python
def plot_bar(chosen_list): 
    #PLot the data distribution for each variable
    for col in chosen_list:
        plt.figure(figsize=(6, 4))
        sns.countplot(x=col, data=survey_data, color='#4e6d91', order=sorted(survey_data[col].dropna().unique()))
        plt.title(f"Distribution of {col}" , fontsize=18)
        plt.xticks(rotation=30, fontsize=16)
        plt.yticks(fontsize=16)
        full_path = os.path.join('../figures/distribution', col)
        plt.savefig(full_path, dpi=300, bbox_inches='tight', pad_inches=0.05)
        plt.tight_layout()
        plt.show()
```

## 1. Data selection

```python
#Clustering columns
selected_columns=[ 'Trust Prime Minister' , 'Trust Government Representative', 'Trust Family and Friends', 'Trust General Media', 'Trust Social Media', 
                  'Household Income', 'Household Savings','Multiple Incomes', 'Age Group', 'Any measure done non-structural', 'Any measure done structural','Any measure plan non-structural', 'Any measure plan structural' , 'Flood Experience'] 
```

## 2. Plotting the variables

```python
#Plotting the data distribution for the variables
plot_bar(selected_columns)
```
